```python
def roundtime(x):
    if(len(str(x))==4):
        if(int(str(x)[0:2]=='23')and int(str(x)[-2::1]>='30')):
            print('here')
            return 0
        if(int(str(x)[-2::1])<=30):
            return int(str(x)[:2]+"00")
            
        if(int(str(x)[-2::1])>30):
            return int(str(x)[:2]+"00")+100
    
    else:
        if(int(str(x)[-2::1])<=30):
            return int(str(x)[:1]+"00")
        if(int(str(x)[-2::1])>30):
            return int(str(x)[:1]+"00")+100
df1['time'] = df1['CRSDepTime'].apply(roundtime)
```
